#### For 2d projections
- we have to compute the projection of the 3d ellipsoid onto the xy, yz, xz plane 
- these projections define a new ellipse with 2 new "eigenvalues" s1, s2
- have to use these 2 eigenvalues to define the smoothing kernel and evaluate the kernel at the cell midpoints

```python
pos = pos[:, :2]
quantities = masses[:, np.newaxis]
extent = np.array([0, boxsize])
gridnum = 300


field, weight = isotropic_kernel_deposition(
                    positions=pos,
                    hsm=hsm,
                    quantities=masses,
                    averaged=[False],
                    extent=extent,
                    gridnum=gridnum,
                    periodic=1)


field_ani, weights_ani = anisotropic_kernel_deposition(
                    positions=pos,
                    hmat=hmat,
                    quantities=masses,
                    averaged=[False],
                    gridnum=gridnum,
                    plane='xy',
                    extent=extent,
                    periodic=1)
```
