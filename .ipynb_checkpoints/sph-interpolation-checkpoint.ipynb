```python
import numpy as np
import pynbody as pn
import matplotlib.pyplot as plt

from sph_modules import sph_interpolator, create_grid_2d, create_grid_3d

%matplotlib inline
%config InlineBackend.figure_format='retina'
```

#### Simulation example: 2d projection

+ velocity fields must be deposited individually (vx, vy, and vz)
+ 2d projections compute the kernel size based only on the 2d coordinates

```python
ds = pn.load('../../../../../shares/feldmann.ics.mnf.uzh/data/FIREbox/production_runs/FB15N128/snapshot_344.hdf5')


boxsize = 15000
NN = 16

pos    = ds.gas['pos']
masses = ds.gas['mass']

mask = np.logical_and(pos[:, -1] > 13500,
                      pos[:, -1] < 15000)
pos  = pos[mask]
masses = masses[mask]

Interpolator = sph_interpolator(masses, boxsize, NN).setup(dim=2, 
                                                           pos=pos[:, :2])
```

```python
nx = 1500
grid_2d = create_grid_2d(nx, nx, boxsize)

# Mass deposit does not make sense -> deposit density instead
# need to say which particles have EOS...
# p = p2m_obj.rho * ds.g['temp'].in_units('K')[mask] #ds.g['rho'].in_units('m_p cm**-3') * ds.g['temp'].in_units('K')

A       = Interpolator.rho #ds.gas['u'][mask] # * p2m_obj.rho
A_grid  = Interpolator.query_field_at_positions(A, grid_2d)
A_grid  = A_grid.reshape((nx, nx))
```

```python
plt.figure()
plt.imshow(np.rot90(np.log10(A_grid)))
plt.axis('off')
plt.show()
```

```python
total_mass_sim = masses.sum()
total_mass_grid = A_grid.mean() * 15000**2

print(total_mass_sim / total_mass_grid)
```

#### Simulation example 3d

```python
ds = pn.load('../../../../../shares/feldmann.ics.mnf.uzh/data/FIREbox/production_runs/FB15N128/snapshot_344.hdf5')


boxsize = 15000
nx = 200
NN = 8

pos    = ds.gas['pos']
masses = ds.gas['mass']

Interpolator = sph_interpolator(masses, boxsize, NN).setup(dim=3, pos=pos)
```

```python
# Mass deposit does not make sense -> deposit density instead
# need to say which particles have EOS...
grid_3d = create_grid_3d(nx, nx, nx, boxsize)

A       = Interpolator.rho #p #ds.gas['ke'] # p2m_obj.prho
A_grid  = Interpolator.query_field_at_positions(A, grid_3d)
A_grid  = A_grid.reshape((nx, nx, nx))
```

```python
plt.figure()
plt.imshow(np.rot90(np.log10(A_grid.sum(axis=-1))))
plt.axis('off')
plt.show()
```

```python
total_mass_sim = masses.sum()
total_mass_grid = A_grid.mean() * 15000**3

print(total_mass_sim / total_mass_grid)
```
